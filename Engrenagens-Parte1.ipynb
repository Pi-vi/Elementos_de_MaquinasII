## Teoria do Dente de Engrenagem

→ O meio mais fácil de transferir movimento rotatório de um eixo a outro é com um par de cilindros rodando.  
→ As principais deficiências são a capacidade relativamente baixa de torque e a possibilidade de escorregamento.

**A lei fundamental do engrenamento**

→ Conceitualmente, dentes de qualquer forma prevenirão o escorregamento bruto. No entanto, algumas forma violam a lei fundamental do engrenamento.

→ A razão de engrenagens deve manter-se constante durante o engrenamento. A razão de velocidade angular $m_v$ é igual a razão do raio primitivo (de referência) da engrenagem de entrada para aquela engrenagem de saída:

$m_v = \dfrac{\omega_1}{\omega_2} = \pm \dfrac{r_2}{r_1}$

A razão de torque ou ganho mecânico $m_A$ é o recíproco da razão de velocidade $m_v$;

$m_A = \dfrac{T_1}{T_2} = \dfrac{1}{m_v} = \dfrac{\omega_2}{\omega_1} = \pm \dfrac{r_1}{r_2}$


- Um par de engrenagens é essencialmente um dispositivo de troca de torque por velocidade e vice-versa. 

A razão de engrenamento $m_G$ é entendida como a magnitude da razão de velocidades ou de torques, qualquer deles que seja > 1:

$m_G = |m_v|$  ou $m_G = |m_A|$, para $m_g \geq 1$

A fim de que a lei fundamental de engrenamento seja verdadeira, os contornos nos dentes devem ser conjugados um ao outro.

**A forma involuta do dente**

A involuta de um círculo é a curva que pode ser gerada desenrolando-se uma linha esticada de um cilindro.

![](figuras/fig7-1.png)

- A linha está sempre tangente ao círculo de base
- O centro de curvatura da involuta está sempre em um ponto de tangência da linha com o círculo de base
- Uma tangente à involuta é sempre normal à linha, que é o raio instantâneo de curvatura da curva.

**Linha de pressão**

![](figuras/fig7-2.png)

- Há uma normal comum a ambas as curvas involutas do dente no ponto de contato.
- A normal comum, que é também a linha de ação, sempre passa pelo ponto de referência (primitivo) sem importar onde no engrenamento os dois estão se contatando.
- O ponto primitivo tem a mesma “velocidade de linha de referência (primitiva)”.

**Ângulo de pressão**

O ângulo de pressão $\phi$ de um par de engrenagens é definido como o ângulo entre a linha de ação (normal comum) e a direção da velocidade no ponto de referência (primitivo) tal que a linha de açõa seja rodada de um ângulo $\phi$ em graus na direção de rotação da engrenagem movida.

**Geometria do engrenamento**

Lei fundamental do engrenamento: a normal comum do perfil de dentes, em todos os pontos de contato durante o engrenamento, deve sempre passar por um ponto fixo na linha de centro das engrenagens, chamado ponto de referência. 


- A distância ao longo da linha de ação entre esses pontos dentro do engrenamento é chamada comprimento de ação $Z$

$Z = \sqrt{(r_p + a_p)^2 - (r_p cos \phi)^2} + \sqrt{(r_g + a_g)^2 - (r_g cos \phi)^2} - C sen \phi$

$r_p$ e $r_g$ são os raios dos círculos primitivos do pinhão e da coroa.

$a_p$ e $a_g$ são os adendos

$C$ é a distância entre centros

![](figuras/fig7-3.png)

**Folga de engrenamento**

A folga é definida como o intervalo entre dentes engrenados medida ao longo da circunferência do círculo primitivo

→ pode ocorrer por redução da espessura do dente, ou por afastamento dos centros

## Nomenclatura do Dente de Engrenagem

![](figuras/fig7-4.png)

Passo circular:
$p_c = \dfrac{\pi d}{N}$

$d$ = diâmetro primitivo
$N$ = número de dentes

Passo circular de base:
$p_b = p_c  cos \phi$

→ Passo diametral:
$p_d = \dfrac{N}{d}$ [polegadas]

→ Módulo:
$m = \dfrac{d}{N}$ [milímetros]

As engrenagens métricas não são intercambiáveis com as engrenagens estadunidenses.

## Trem de Engrenagens

→ Coleção de duas ou mais engrenagens acolpladas

→ Um par de engrenagens é, portanto, a forma mais simples de um trem de engrenagem e usualmente está limitada a uma razão de cerca de 10:1 


- Simples
- Composta 
- Epicíclicos

**Trem de engrenagens simples**


- cada eixo carrega somente uma engrenagem

![](figuras/fig7-6.png)

A razão de velocidade, ou razão de trem, de um par de engrenagem é dado pela equação:

$m_v = \left(\dfrac{-N_2}{N_3} \right) \left(\dfrac{-N_3}{N_4} \right)\left(\dfrac{-N_4}{N_5} \right)\left(\dfrac{-N_5}{N_6} \right) = +\dfrac{N_2}{N_6}$


**Trem de engrenagens composto**

→ Para obter uma razão de  velocidade maior que 10:1 com engrenagens é necessário compor o trem.

→ Em um trem composto, pelo menos um eixo tem mais do que uma engrenagem.

![](figuras/fig7-7.png)


A razão de velocidade do trem composto é:

$m_v = \left(-\dfrac{N_2}{N_3}\right) \left(-\dfrac{N_4}{N_5}\right)$
